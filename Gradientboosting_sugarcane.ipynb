```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Load the data from CSV
csv_path = '/content/drive/MyDrive/Sugarcaneidentification/traintestdataset.csv'
data = pd.read_csv(csv_path)
```

```python
# Extract features (NDVI, NDWI, SAVI, Red_Band, Blue_Band, Green_Band) and target (Class_ID)
features = data[['NDVI', 'NDWI', 'SAVI', 'Red_Band', 'Blue_Band', 'Green_Band']]
target = data['Class_ID']
```

```python
# Extract features (NDVI, NDWI, SAVI, Red_Band, Blue_Band, Green_Band) and target (Class_ID)
features = data[['NDVI', 'NDWI', 'SAVI', 'Red_Band', 'Blue_Band', 'Green_Band']]
target = data['Class_ID']
```

```python
# Initialize the Gradient Boosting Classifier
clf = GradientBoostingClassifier()
```

```python
# Split the data into training and testing sets (80% for training, 20% for testing)
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Train the model on the training data
clf.fit(X_train, y_train)
```

```python
# Make predictions on the testing data
y_pred = clf.predict(X_test)
```

```python
# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Calculate the confusion matrix
confusion_mat = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
print(class_report)
print("Confusion Matrix:")
print(confusion_mat)
```

```python
# Optionally, get the confidence score for the model
confidence_score = clf.score(X_test, y_test)
print("Confidence Score:", confidence_score)
```
